6. Create a data frame with only age, education, and occupation by using subsetting.

```python
# This code creates a subset of the inc_data data frame consisting of age, education, and occupation
inc_data_1 = inc_data[["age", "education", "occupation"]]
```

```python
# This code displays the first 5 rows of the new data set
inc_data_1.head()
```

7. Create a histogram of age with bin size of 20.

```python
# This code creates a histogram of the age column
plt.figure(figsize= (8,10))
plt.hist(inc_data_1["age"], bins=20, color="green", edgecolor= "black")
plt.title("Age", fontsize= 40)
plt.xlabel("Age in Years", fontsize = 25)
plt.ylabel("Total in groups", fontsize = 25)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
sns.despine()
plt.show()

```

```python
# This code creates a boxplot of age and education

inc_data_1.boxplot(column="age", by="education", figsize=(25,10))
plt.title("Age VS Education", fontsize = 40, weight = "bold" )
plt.xlabel("Education", fontsize = 40, weight = "bold")
plt.ylabel("Age", fontsize = 40)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 30)
plt.show()
```

8. Create a function to strip the white space characters.

```python
# This code creates a function to remove white spaces
def strip_whitespace(s):
    return s.strip()
```

9. Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column and drop the new column.

```python


import warnings
warnings.filterwarnings("ignore")
inc_data_1["education_clean"] = inc_data_1["education"].apply(strip_whitespace)
inc_data_1["education"] = inc_data_1["education_clean"]
inc_data_1.drop(labels = ["education_clean"], axis = 1, inplace = True)


inc_data_1["occupation_clean"] = inc_data_1["occupation"].apply(strip_whitespace)
inc_data_1["occupation"] = inc_data_1["occupation_clean"]
inc_data_1.drop(labels = ["occupation_clean"], axis = 1, inplace = True)
```
