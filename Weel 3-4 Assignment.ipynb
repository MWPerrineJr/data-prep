8. Create a function to strip the white space characters.

```python
# This code creates a function to remove white spaces
def strip_whitespace(s):
    return s.strip()
```

9. Use the apply method to apply this function to all the columns with string values, create a new column, copy the values from this new column to the old column and drop the new column.

```python


import warnings
warnings.filterwarnings("ignore")
inc_data_1["education_clean"] = inc_data_1["education"].apply(strip_whitespace)
inc_data_1["education"] = inc_data_1["education_clean"]
inc_data_1.drop(labels = ["education_clean"], axis = 1, inplace = True)


inc_data_1["occupation_clean"] = inc_data_1["occupation"].apply(strip_whitespace)
inc_data_1["occupation"] = inc_data_1["occupation_clean"]
inc_data_1.drop(labels = ["occupation_clean"], axis = 1, inplace = True)
```

# 3. Create a Series and Practice Basic Arithmetic Steps
  

1. Create Series

```python
# This code creates the data for the series
data= [7.3, -2.5, 3.4, 1.5]
index = ["a", "c", "d", "e"]

data_2 = [-2.1, 3.6, -1.5, 4, 3.1]
index_2 = ["a","c", "e", "f", "g"]
```

```python
# This code creates the series based on the given variables
series_1 = pd.Series(data, index)
series_2 = pd.Series(data_2, index_2)
print(series_1, series_2)
```

```python
# This code adds the two series
add = series_1 + series_2
add
```

```python
# This code subtracts the two series
subtract = series_1 - series_2
subtract
```

```python
# This code multiplies the two series
multiply = series_1 * series_2
multiply
```

```python
# This code divides the two series
divide = series_1 / series_2
divide
```
